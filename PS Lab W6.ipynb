```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb

N = 100
np.random.seed(42)
visits = np.random.randint(0, 11, N)

unique, counts = np.unique(visits, return_counts=True)
pmf = counts / N
cdf = np.cumsum(pmf)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.bar(unique, pmf, color='blue', alpha=0.7)
plt.xlabel("Visitors")
plt.ylabel("Probability")
plt.title("PMF")

plt.subplot(1, 2, 2)
plt.step(unique, cdf, where="mid", color='red', alpha=0.7)
plt.xlabel("Visitors")
plt.ylabel("Cumulative Probability")
plt.title("CDF")

plt.tight_layout()
plt.show()

N_students = 100
p = 0.7
k = 70

def binomial_pmf(n, k, p):
    return comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

prob_exact_k = binomial_pmf(N_students, k, p)
prob_at_least_k = sum(binomial_pmf(N_students, i, p) for i in range(k, N_students + 1))
expected_pass = N_students * p

print(f"P(X = {k}): {prob_exact_k:.4f}")
print(f"P(X ≥ {k}): {prob_at_least_k:.4f}")
print(f"Expected Pass: {expected_pass}")
```

```python
import scipy.stats as stats
p = 0.7
N = 100
k = 60
prob_exactly_60 = stats.binom.pmf(k, N, p)
print(f"Probability of exactly 60 students passing: {prob_exactly_60}")
k = 75
prob_at_least_75 = 1 - stats.binom.cdf(k - 1, N, p)
print(f"Probability of at least 75 students passing: {prob_at_least_75}")
expected_value = N * p
print(f"Expected number of students passing: {expected_value}")
```
